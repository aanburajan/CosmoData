#Cosmo Implementation

    youtubeLinks contains the dataset of youtube.com/watch links
    we are going to append a column next to the csv .
    using pafy we are going to scrap each video link for the meta data like title, description, and tags

```python
import csv


with open('youtubeLinks.csv','r') as csvInput:
    with open('newLinks.csv','w') as csvOutput:
        writer = csv.writer(csvOutput, lineterminator = '\n')
        reader = csv.reader(csvInput)
        
        all = []
        row = reader.next()
        row.append('title')
        row.append('description')
        row.append('tags')
        all.append(row)
        for row in reader:
            try:
                tempUrl = row[0]
                tempVideo = pafy.new(tempUrl)
                row.append(tempVideo.title)
                row.append(tempVideo.description)
                row.append(tempVideo.keywords)
                
            except IOError:
                row.append('private')
                row.append('private')
                row.append('private')
            except Exception:
                row.append('err')
                row.append('err')
                row.append('err')
            all.append(row)

            
```

So at this point, all[][9:11] contains the mined data
9 - title

```python

forCSV = []
for i in range(1,len(all)):
    all[i][11] = ' '.join(all[i][11])
    tempBag = all[i][9] + ' ' + all[i][10] + ' ' + all[i][11]
    tempBag = tempBag.replace('\n','')
    tempBag = tempBag.replace('\r','')
    tempBag = tempBag.replace('\'s','')
    tempBag = tempBag.replace('\'','\'')
    tempBag = tempBag.replace('\\u','')
    tempBag = tempBag.replace('\"','')
    tempBag = tempBag.replace('.','')
    tempBag = tempBag.replace('err','')
    tempBag = tempBag.replace('private','')
    tempBag = tempBag.replace('?','')
    tempBag = tempBag.replace('!','')
    tempBag = tempBag.replace('(','')
    tempBag = tempBag.replace(')','')
    tempBag = tempBag.replace('http',' http')
    tempBag = tempBag.replace('-','')
    tempBag = tempBag.replace('  ',' ')
    
    all[i][10]
    all[i][9]
    all[i].append(tempBag)
    forCSV.append(tempBag)
    
    
```

We have cleaned the data to an extend and we are ready to do some NLP clustering
We will be using the gensim library
    https://radimrehurek.com/gensim/tut1.html#from-strings-to-vectors
    
We are going to follow procedure layed out here for lsa :
    https://www.tacoma.uw.edu/sites/default/files/global/documents/institute_tech/mpaul.pdf

var texts contains the tokens used for our dictionary

var dictionary contains all the unique tokens

```python
from gensim import corpora, models, similarities

stoplist = set('for a of the and to in on is our it & oh with or e r'.split())
texts = [[term for term in instance.lower().split() if term not in stoplist] for instance in forCSV]

from collections import defaultdict
frequency = defaultdict(int)
for text in texts:
    for term in text:
        frequency[term] += 1

texts = [[token for token in text if frequency[token] >1] for text in texts]

```

```python
dictionary = corpora.Dictionary(texts)
dictionary.save('gensim/youtube.dict')
corpus = [dictionary.doc2bow(text) for text in texts]
corpora.MmCorpus.serialize('gensim/youtubelink.mm', corpus)
```
